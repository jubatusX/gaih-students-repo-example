```
import random
kelimeler=["İSTATİSTİK","MATEMATİK","FİZİK","MÜZİK","SPOR"]
isim=input("Lütfen isminizi giriniz: ")
seckelime=random.choice(kelimeler)

uzunluk=len(seckelime)
gorsel=list("_" * uzunluk)
tahminler=[]

can=7

print("#BİLGİ","\n","[1]Her el sadece 1 harf tahmin edebilirsiniz.","\n","[2]7 can hakkınız var, her yanlış tahminde 1 canınız gitmektedir.","\n","[3]Dilerseniz direkt tahminde bulunabilirsiniz.Yanlış bilmeniz halinde direkt oyunu kaybedersiniz."
,"\n","[4]Tahminlerinizde büyük harf kullanınız.(Özellikle türkçe karakterlerde.)")
print("\n","Haydi Oyun Başlasın",isim.upper(),"\n","İYİ ŞANSLAR !\n")
print("\n","Tahmin Edeceğin Kelime : ", " ".join(gorsel),"\n")

while can > 0:
  veri=input("[?] Bir Harf Tahmininde Bulununuz : ")
  tahmin=veri.upper()

  if tahmin in tahminler:
    print("Önceki Harf Tahminlerinize Dikkat Ediniz. \n [HATIRLATMA]=> Daha önceki tahmin harfleri :",tahminler)
    continue
  
  elif len(tahmin) > 1 or len(tahmin) == 0:
    print("[UYARI] Bir Harf Giriniz !")
    continue
    
  elif tahmin not in seckelime:
    can -=1
    print("[✗] Yanlış tahmin",can," ❣ kaldı.")
    tahminler +=tahmin
  else:
    for i in range(len(seckelime)):
      if tahmin == seckelime[i]:
        gorsel[i]=tahmin
    tahminler +=tahmin
    print("[✓] Doğru Tahmin , Yuppi...")
    print(" ".join(gorsel))

  istek =input("[^.^] Direkt tahminde bulunmak istermisiniz (E/.) : ")
  
  if istek.upper() == "E":
    direktahmin=input("[?] Tahmininiz nedir ? : ")
    if direktahmin.upper() == seckelime:
      print("Tebrikler Kelimeyi Doğru Tahmin Ettiniz.") 
      break
    else:
      print("Yanlış Tahminde Bulundunuz.","\n","[☠] KAYBETTİNİZ.")
      break
  if can == 0:
    print("Canınız kalmadı. [☠] KAYBETTİNİZ")
    break
```
